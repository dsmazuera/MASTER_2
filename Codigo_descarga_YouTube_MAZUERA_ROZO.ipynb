```python
videos = [
    "https://www.youtube.com/watch?v=yaA1_92D_Lo",
    "https://www.youtube.com/watch?v=Fb0w5gEO9bc",
    "https://www.youtube.com/watch?v=bMbRXjR4EFM",
    "https://www.youtube.com/watch?v=AeZHFKdanG8",
    "https://www.youtube.com/watch?v=O4XYHdIsZQY",
    "https://www.youtube.com/watch?v=YvvPeimqLQI",
    "https://www.youtube.com/watch?v=IFd9u69HxeU",
    "https://www.youtube.com/watch?v=xJaW69rv0z0",
]
len(videos)
```

```python
from youtube_comment_downloader import *
import pandas as pd

def get_youtube_comments(url, max_comments=None):

    downloader = YoutubeCommentDownloader()
    comments = downloader.get_comments_from_url(url, sort_by=SORT_BY_POPULAR)

    data = []
    for i, comment in enumerate(comments):
        if max_comments is not None and i >= max_comments:
            break
        data.append({
            "video_url": url,
            "author": comment.get("author"),
            "text": comment.get("text"),
            "time": comment.get("time"),
            "votes": comment.get("votes"),
        })
    return pd.DataFrame(data)
```

```python
dfs = []

for url in videos:
    print(f"Descargando comentarios de: {url}")
    df = get_youtube_comments(url, max_comments=10000)
    print(f"   Comentarios obtenidos: {len(df)}\n")
    dfs.append(df)

df_yt = pd.concat(dfs, ignore_index=True)
print("Total de comentarios:", len(df_yt))
df_yt.head()
```

```python
df_yt.to_csv("comentarios_centros_poblados_youtube_sin_api.csv", index=False)
"Guardado como comentarios_centros_poblados_youtube_sin_api.csv"
```
